# **<u>Description of Titanic Dataset</u>**

Each row represents data for a specific entity or state.

Description of the columns in the dataset is as follows:

**<u>PassengerId</u>**: An identifier for each passenger.

**<u>Survived</u>**: This column indicates whether a passenger survived (1) or did not survive (0).

**<u>class</u>**: The passenger class (ticket class) indicates whether the passenger was in the first class (1), second class (2), or third class (3).

**<u>Name</u>**: The name of the passenger.

**<u>Sex</u>**: The gender of the passenger (male or female).

**<u>Age</u>**: The age of the passenger in years.

**<u>SibSp</u>**: The number of siblings or spouses the passenger had aboard the Titanic.

**<u>Parch</u>**: The number of parents or children the passenger had aboard the Titanic.

**<u>Ticket</u>**: The ticket number of the passenger.

**<u>Fare</u>**: The fare paid by the passenger for the ticket.

**<u>Cabin</u>**: The cabin number where the passenger stayed (contains missing values).

**<u>Embarked</u>**: The port of embarkation, i.e., the location where the passenger boarded the Titanic (C = Cherbourg, Q = Queenstown, S = Southampton).

# **<u>Outlier Detection</u>:-**

&bull; **<u>First Quartile(Q1)</u>** :- Q1 is the data point that separates the lowest 25% of the data from the rest of the data when the data is sorted in ascending order.

&bull; **<u>Second Quartile(Q2)</u>** :- It is also simply called as 'Median'. It divides the data into two equal halves, with 50% of the data points falling below it and 50% falling above it.

&bull; **<u>Third Quartile(Q3)</u>** :- Q3 is the data point that separates the lowest 75% of the data from the highest 25% when the data is sorted in ascending order.

&bull; **<u>Interquartile Range(IQR)</u>** :-The IQR is useful for identifying potential outliers in a dataset. Data points that fall below Q1 - 1.5 * IQR or above Q3 + 1.5 * IQR are considered potential outliers.

&bull;**<u>Mathematically, the IQR is defined as:</u>** IQR=Q3âˆ’Q1

```python
Q1_age = dataset['Age'].quantile(0.25)
Q3_age = dataset['Age'].quantile(0.75)
IQR_age = Q3_age - Q1_age
lower_bound_age = Q1_age - 1.5 * IQR_age
upper_bound_age = Q3_age + 1.5 * IQR_age

outliers_age = dataset[(dataset['Age'] < lower_bound_age) | (dataset['Age'] > upper_bound_age)]
```

```python
outliers_age
```

```python
Q1_fare = dataset['Fare'].quantile(0.25)
Q3_fare = dataset['Fare'].quantile(0.75)
IQR_fare = Q3_fare - Q1_fare
lower_bound_fare = Q1_fare - 1.5 * IQR_fare
upper_bound_fare = Q3_fare + 1.5 * IQR_fare

outliers_fare = dataset[(dataset['Fare'] < lower_bound_fare) | (dataset['Fare'] > upper_bound_fare)]
```

```python
outliers_fare
```
